## **Central Tendency**

Different measures of central tendency are : 
1. Arithmetic mean 
2. Median
3. mode 

```python
columns = list(df.columns)
```

```python
#Arithmetic mean  
print("Mean".center(70,'-'))
df.mean(numeric_only=True)
```

```python
#Median of the columns 
print("Median".center(90,'-'))
df.median(numeric_only=True)
```

```python
#Mode of the columns 
print("Mode".center(90,'-'))
df.mode(axis='columns',numeric_only=True)
```

## **Measures of Dispersion**

1. Range
2. Quartile Deviation 
3. Mean Deviation
4. Standard Deviation and Variance

```python
#range of the columns 
print("Range".center(90,'-'))
for i in columns:
    if(df[i].dtype==np.float64 or df[i].dtype==np.int64):
        print("Range of the column",i,":",(df[i].max()-df[i].min()))
```

```python
#Quartile Deviation 
for i in columns:
    if(df[i].dtype==np.float64 or df[i].dtype==np.int64):
        temp=df[i].quantile([0.25,0.50,0.75])
        print("Quartile Deviation of",i,":",((temp[0.75]-temp[0.25])/2))
```

```python
Mu = temp['Sample'].mean()
t =0
for i in temp['Sample'].iteritems():
    t=t+(i[1]-Mu)
t/len(temp['Sample'])
```

```python
for i in columns:
    t=0
    for j in df[i].iteritems():
        t = t + (df[i].mean())
    print("MD of",i,":",(t/len(df[i])))
```

```python
df.mad()
```

```python
#variance 
print("Variance".center(90,'-'))
df.var(numeric_only=True)
```

```python
#standard deviation 
print("Standard Deviation".center(40,'-'))
df.std(numeric_only=True)
```
